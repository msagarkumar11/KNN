```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.feature_names
```

```python
print(digits.data.shape)
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df.head()
```

dataset size: 1797 x 64

```python
df['classes'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['classes'], axis='columns')
y = df.classes
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
knn_scores=[]
k_range=range(1,21)
for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  score=knn.score(X_test, y_test)
  knn_scores.append(score)
print(knn_scores)
```

```python
from matplotlib.cm import rainbow
plt.bar(k_range, knn_scores,width=0.6)
for i in k_range:
    plt.text(i, knn_scores[i-1], f"{knn_scores[i-1]:.2f}")
plt.xticks(k_range)
plt.xlabel('K Values')
plt.ylabel('Scores')
plt.show()
```

```python
max_accuracy=max(knn_scores)
optimal_k=knn_scores.index(max_accuracy)+1
print(optimal_k,max(knn_scores))
```

```python
knn = KNeighborsClassifier(n_neighbors=optimal_k)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
